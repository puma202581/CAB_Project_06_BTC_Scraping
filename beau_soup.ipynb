```python
import requests
import bs4


def scrape_data():
    url = "https://quotes.toscrape.com/"
    response = requests.get(url)
    soup = bs4.BeautifulSoup(response.text, 'html.parser')
    print(soup)
    return soup


data = scrape_data()
```

```python
# Scraping author names from the quotes page
def scrape_authors(soup):
    authors = soup.find_all('small', class_='author')
    author_names = [author.get_text() for author in authors]
    return author_names
author_names = scrape_authors(data)
print("Authors found:" + str(author_names))
```

```python
# Scraping quotes from the quotes page
def scrape_einstein_quotes(soup):
    quotes = soup.find_all('div', class_='quote')
    einstein_quotes = []
    for quote in quotes:
        author = quote.find('small', class_='author').get_text()
        if author == 'Albert Einstein':
            text = quote.find('span', class_='text').get_text()
            einstein_quotes.append(text)
    return einstein_quotes

einstein_quotes = scrape_einstein_quotes(data)
print("Einstein Quotes found:" + str(einstein_quotes))
```

```python
# Scraping quotes from the quotes page
def scrape_all_quotes(soup):
    quotes = soup.find_all('div', class_='quote')
    all_quotes = []
    for quote in quotes:
        author = quote.find('small', class_='author').get_text()
        text = quote.find('span', class_='text').get_text()
        all_quotes.append({'author': author, 'quote': text})
    return all_quotes

all_quotes = scrape_all_quotes(data)
print("All Quotes found:", all_quotes)
```

```python
# Create an SQL table to store the quotes and authors

def create_quotes_table():
    import sqlite3
    conn = sqlite3.connect('quotes.db')
    cursor = conn.cursor()
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS quotes (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            author TEXT NOT NULL,
            quote TEXT NOT NULL
        )
    ''')
    conn.commit()
    conn.close()
create_quotes_table()

# Function to insert quotes into the database
def insert_quote(author, quote):
    import sqlite3
    conn = sqlite3.connect('quotes.db')
    cursor = conn.cursor()
    cursor.execute('''
        INSERT INTO quotes (author, quote) VALUES (?, ?)
    ''', (author, quote))
    conn.commit()
    conn.close()

# Insert the scraped quotes into the database
for quote in all_quotes:
    insert_quote(quote['author'], quote['quote'])

# Function to retrieve quotes from the database
def retrieve_quotes():
    import sqlite3
    conn = sqlite3.connect('quotes.db')
    cursor = conn.cursor()
    cursor.execute('SELECT * FROM quotes')
    rows = cursor.fetchall()
    conn.close()
    return rows

# Retrieve and print the quotes from the database
quotes_from_db = retrieve_quotes()
for row in quotes_from_db:
    print(f"Author: {row[1]}, Quote: {row[2]}")

# Function to delete a quote from the database
def delete_quote(quote_id):
    import sqlite3
    conn = sqlite3.connect('quotes.db')
    cursor = conn.cursor()
    cursor.execute('DELETE FROM quotes WHERE id = ?', (quote_id,))
    conn.commit()
    conn.close()

# # Example usage of delete_quote function
# delete_quote(1)  # Replace 1 with the actual ID of the quote you want to delete
```
